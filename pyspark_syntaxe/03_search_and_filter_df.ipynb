```python
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("searchDF").getOrCreate()
spark
```

```python
path = 'datasets/'

fifa = spark.read.csv(path+'fifa19.csv', inferSchema=True, header=True)
```

```python
fifa.toPandas()
```

```python
fifa.printSchema()
```

```python
from pyspark.sql.functions import *
```

```python
fifa.select(['Nationality', 'Name', 'Age', 'Photo']).show(5, False)
```

```python
fifa.select(['Name', 'Age']).orderBy(fifa['Age']).show(5)
```

```python
fifa.select(['Name', 'Age']).orderBy(fifa['Age'].desc()).show(5)
```

```python
fifa.select(['Name', 'Club']).where(fifa["Club"].like("%Barcelona%")).show(5, False)
```

```python
fifa.select('Photo', fifa["Photo"].substr(-4, 4)).show(5)
```

```python
fifa[fifa["Club"].isin("FC Barcelona", "Juventus")].limit(5).toPandas()
```

```python
fifa.select(['Name', 'Club']).where(fifa["Name"].startswith("L.")).where(fifa["Name"].endswith("i")).limit(5).toPandas()
```
